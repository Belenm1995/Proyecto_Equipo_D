```python
# Datos Caldes de Montbui 2024/2025
data = {}

url = "https://api.weatherbit.io/v2.0/history/daily?lat=41.63333&lon=2.16667&start_date=2024-11-15&end_date=2025-11-15&key=9eff7a274b02431bbc87d0660dbabd60"
response = requests.get(url)

response
```

```python
data.update(response.json())
```

```python
pprint(data)
```

```python
ciudad = data["city_name"]
ciudad
```

```python
datos = {}
city_name = []
datetime = []
ts = []
pres = []
slp = []
wind_spd = []
wind_gust_spd = []
max_wind_spd = []
wind_dir = []
max_wind_dir = []
max_wind_spd_ts = []
temp = []
max_temp = []
min_temp = []
max_temp_ts = []
min_temp_ts = []
rh = []
dewpt = []
clouds = []
precip = []
precip_gpm = []
snow = []
snow_depth = []
solar_rad = []
t_solar_rad = []
ghi = []
t_ghi = []
max_ghi = []
dni = []
t_dni = []
max_dni = []
dhi = []
t_dhi = []
max_dhi = []
max_uv = []

for x in data["data"]:
    city_name.append(ciudad)
    datetime.append(x["datetime"])
    ts.append(x["ts"])
    pres.append(x["pres"])
    slp.append(x["slp"])
    wind_spd.append(x["wind_spd"])
    wind_gust_spd.append(x["wind_gust_spd"])
    max_wind_spd.append(x["max_wind_spd"])
    max_wind_spd_ts.append(x["max_wind_spd_ts"])
    wind_dir.append(x["wind_dir"])
    temp.append(x["temp"])
    max_temp.append(x["max_temp"])
    min_temp.append(x["min_temp"])
    max_temp_ts.append(x["max_temp_ts"])
    min_temp_ts.append(x["min_temp_ts"])
    rh.append(x["rh"])
    dewpt.append(x["dewpt"])
    clouds.append(x["clouds"])
    precip.append(x["precip"])
    precip_gpm.append(x["precip_gpm"])
    snow.append(x["snow"])
    snow_depth.append(x["snow_depth"])
    solar_rad.append(x["solar_rad"])
    t_solar_rad.append(x["t_solar_rad"])
    ghi.append(x["ghi"])
    t_ghi.append(x["t_ghi"])
    max_ghi.append(x["max_ghi"])
    dni.append(x["dni"])
    t_dni.append(x["t_dni"])
    max_dni.append(x["max_dni"])
    dhi.append(x["dhi"])
    t_dhi.append(x["t_dhi"])
    max_dhi.append(x["max_dhi"])
    max_uv.append(x["max_uv"])

    for llave, valor in [("Ciudad", city_name),
                     ("Fecha", datetime), 
                     ("Fecha (UNIX)", ts), 
                     ("Presin atmosfrica", pres), 
                     ("Presin Atm. sobre Nvl. del mar", slp), 
                     ("Media de la vel. del viento (m/s)", wind_spd), 
                     ("Vel. rfaga de viento max.", wind_gust_spd),
                     ("Mx. vel. del viento /2min", max_wind_spd), 
                     ("Momento (UNIX) de la vel. mx. del veinto", max_wind_spd_ts), 
                     ("Media de la direccin del viento", wind_dir), 
                     ("Media de la temperatura", temp), 
                     ("Temperatura mxima", max_temp), 
                     ("Temperatura mnima", min_temp), 
                     ("Momento temperatura mxima (UNIX)", max_temp_ts), 
                     ("Momento temperatura mnima (UNIX)", min_temp_ts), 
                     ("Media humedad", rh), 
                     ("Media punto de roco", dewpt), 
                     ("Media de % de nubes en la zona", clouds), 
                     ("Acumulacin de precipitaciones (l/m2)", precip), 
                     ("Acumulacin de precipitaciones (l/m2) (NASA)", precip_gpm), 
                     ("Acumulacin de nieve (l/m2)", snow), 
                     ("Profundidad de la neive", snow_depth), 
                     ("Media de radiacin solar (W/m2)", solar_rad), 
                     ("Total de radiacin solar (W/m2)", t_solar_rad), 
                     ("Media de la radiacin solar horizontal", ghi), 
                     ("Total de la radiacin solar horizontal", t_ghi), 
                     ("Mximo de la radiacin solar horizontal", max_ghi), 
                     ("Media de la radiacin solar concentrada", dni), 
                     ("Total de la radiacin solar concentrada", t_dni), 
                     ("Mximo de la radiacin solar concentrada", max_dni), 
                     ("Media de la radiacin solar horizontal difusa", dhi), 
                     ("Total de la radiacin solar horizontal difusa", t_dhi), 
                     ("Mximo de la radiacin solar horizontal difusa", max_dhi),
                     ("ndice mximo de rayos ultravioleta", max_uv)]:
        datos[llave] = valor

caldes =  pd.DataFrame(datos)

caldes.head()

caldes.to_csv("caldes_2024-11-15_2025-11-15.csv")
```
